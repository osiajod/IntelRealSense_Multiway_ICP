# Read the ply mesh

```
import open3d as o3d

# import a sample ply
pcl1 = o3d.io.read_triangle_mesh("rotary_room_scanner/rotary_rsrc/clouds2/cloud1607832794.116435.ply")

pcl1_ply = o3d.io.read_point_cloud("rotary_room_scanner/rotary_rsrc/clouds2/cloud1607832794.116435.ply")
```

# Remove noise in the point cloud

```
# First down sample the point cloud

voxel_down_pcd = pcl1_ply.voxel_down_sample(voxel_size = 0.02)
cl, ind = voxel_down_pcd.remove_statistical_outlier(nb_neighbors=20,
                                                    std_ratio=2.0)
```
