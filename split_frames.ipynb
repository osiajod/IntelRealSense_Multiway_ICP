```python
import cv2                               # OpenCV
import numpy as np                       # numpy
import os
import requests
import pickle
import glob   # filename and path management for file I/O


# Use this line to make matplotlib plot inline (only need to call it once when importing matplotlib)
%matplotlib inline

import matplotlib.pyplot as plt


import av
import pims
from PIL import Image
import plotly.express as px
import plotly.graph_objects as go
```

```python
def generate_frames(video_path="./rotary_rsrc/15min_cleanroom1.mp4", output_path="./rotary_rsrc/cleanroom1_frames", step=50, start_frame=0):
    
    v = pims.Video(video_path)
    for num, i in enumerate(v[::step]):
        im = Image.fromarray(v[num*step])
        im.save(f'{output_path}/frame{num+start_frame}.png')
#     print(len(v))
```

```python
generate_frames()
```
